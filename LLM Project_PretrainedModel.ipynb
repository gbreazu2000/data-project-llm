```python
#Importing pre-trained model pipeline
from transformers import pipeline

pipe = pipeline("text-classification", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english", max_length=512, truncation=True, padding=True)
```

```python
sample = test_data.shuffle()[:100]
prediction = pipe(sample['text'])
```

```python
prediction = [0 if test_data['label'] == 'neg' else 1 for label in prediction]
```

```python
from sklearn.metrics import classification_report

print(classification_report(sample['label'], prediction))
```
